```python
df = df.drop_duplicates(subset = ["Movie_Title"])

df = df.reset_index(drop=True)

df['Director_Facebook_Likes'] = df['Director_Facebook_Likes'].str.replace('"', '', regex=False)

exclude_column = 'Movie_Title'

numeric_columns = df.select_dtypes(include='number').columns
numeric_columns = numeric_columns.drop(exclude_column, errors='ignore')

df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].mean())

df = df.drop([9, 10]).reset_index(drop=True)

df
```

```python
summary = df.describe()
summary
```

```python
import matplotlib.pyplot as plt
```

```python
top_revenue_movies = df.sort_values(by="Gross", ascending = False).head(10)

plt.figure(figsize=(10, 6))
bars = plt.barh(top_revenue_movies["Movie_Title"], top_revenue_movies["Gross"], color = "skyblue")
plt.xlabel("Gross Revenue")
plt.ylabel("Top 10 Movies by Gross Revenue(in billions)")
plt.title("Top 10 Movies by Gross Revenue")
plt.gca().invert_yaxis()
for bar, value in zip(bars, top_revenue_movies["Gross"]):
    plt.text(value, bar.get_y() + bar.get_height() / 2, f"{value:,.0f}", va = "center")
plt.show()
```
